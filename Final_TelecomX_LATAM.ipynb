##Manejo de inconsistencias

```python
# Copia del dataframe para no modificar el original
df_tl = df_norm.copy()
```

```python
# Eliminar columnas no útiles
df_tl = df_tl.drop(columns=["customerID"])
```

```python
# Eliminar filas con Churn vacío
df_tl = df_tl.dropna(subset=['Churn'])
```

```python
# Codificación de columnas binarias
columnas_binarias = ["Churn", "Partner", "Dependents", "PhoneService", "PaperlessBilling", "gender"]
mapeo_binario = {"No": 0, "Yes": 1, "Male": 0, "Female": 1, "": np.nan}

for col in columnas_binarias:
    df_tl[col] = df_tl[col].map(mapeo_binario)
```

```python
# SeniorCitizen como booleano
df_tl["SeniorCitizen"] = df_tl["SeniorCitizen"].astype(bool)
```

```python
# Columnas multicategoría
columnas_multicategoricas = [
    "MultipleLines", "InternetService", "OnlineSecurity",
    "OnlineBackup", "DeviceProtection", "TechSupport",
    "StreamingTV", "StreamingMovies", "Contract", "PaymentMethod"
]

for col in columnas_multicategoricas:
    df_tl[col] = df_tl[col].astype("category")
```

```python
# Conversión de columnas numéricas
df_tl["Total"] = pd.to_numeric(df_tl["Total"], errors="coerce")
df_tl['Total'] = df_tl['Total'].fillna(0.0)


df_tl["Monthly"] = pd.to_numeric(df_tl["Monthly"], errors="coerce")
df_tl["tenure"] = pd.to_numeric(df_tl["tenure"], errors="coerce")
```

```python
# Verificación final de tipos de datos
df_tl.dtypes
```

```python
# Verificación rápida
df_tl.isna().sum()
```

DISTRIBUCIÓN DE LA EVASIÓN

```python
# Conteo de clientes por estado de evasión
evasión_counts = df_tl["Evasion"].value_counts(dropna=False, normalize=True) * 100

# Crear figura con 1 fila y 2 columnas
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Gráfico de barras
sns.barplot(x=evasión_counts.index.astype(str), y=evasión_counts.values,
            palette="viridis", ax=axes[0])
axes[0].set_title("Distribución de Evasión (%)")
axes[0].set_ylabel("Porcentaje")
axes[0].set_xlabel("Evasión (0=No, 1=Sí, NaN=Desconocido)")

# Gráfico circular
axes[1].pie(evasión_counts.values,
            labels=evasión_counts.index.astype(str),
            autopct='%1.1f%%',
            colors=sns.color_palette("viridis", len(evasión_counts)))
axes[1].set_title("Proporción de Clientes que Evaden")

# Ajustar espacios
plt.tight_layout()
plt.show()
```

RECUENTO DE EVASIÓN POR VARIABLES CATEGÓRICAS

```python
# para mantener lo hecho anteriormente en la transformación de los datos
df_plot = df_tl.copy()

# Mapear valores de género si son 0 y 1 para un mejor entendimient de la gráfica
if set(df_plot["Genero"].unique()) <= {0, 1}:
    df_plot["Genero"] = df_plot["Genero"].map({0: "Hombre", 1: "Mujer"})

# Variables categóricas a graficar
categoricas = [
    "Genero", "Pareja", "Dependientes",
    "Multiples_Lineas", "Servicio_Internet",
    "Tipo_Contrato", "Metodo_Pago"
]

# Calcular límite máximo común
max_val = 0
for col in categoricas:
    conteos = df_plot.groupby([col, "Evasion"]).size()
    max_val = max(max_val, conteos.max())

# Crear subplots con mismo límite
n_cols = 2
n_rows = math.ceil(len(categoricas) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(categoricas):
    sns.countplot(data=df_plot, x=col, hue="Evasion", palette="Set2", ax=axes[i])
    axes[i].set_title(f"Evasión por {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Cantidad de Clientes")
    axes[i].legend(title="Evasión", labels=["No", "Sí"])
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_ylim(0, max_val * 1.1)  # mismo límite para todos

# Ocultar ejes vacíos si los hay
for j in range(len(categoricas), len(axes)):
    fig.delaxes(axes[j])

# Ajustar espaciado vertical y horizontal
plt.subplots_adjust(hspace=0.9, wspace=0.3)

plt.show()
```

FACTURA DIARIA SEGÚN EVASIÓN Y CANTIDAD DE SERVICIOS CONTRATADOS VS EVASIÓN


```python

servicios_cols = [
    "Multiples_Lineas", "Seguridad_EnLinea", "Respaldo_EnLinea",
    "Proteccion_Dispositivo", "Soporte_Tecnico",
    "Streaming_TV", "Streaming_Peliculas"
]

df_tl_servicios_bin = df_tl[servicios_cols].apply(lambda x: x.astype(str).str.lower().isin(["yes", "sí", "1"]).astype(int))
df_tl["Cantidad_Servicios"] = df_tl_servicios_bin.sum(axis=1)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot Factura Diaria
sns.boxplot(data=df_tl, x="Evasion", y="Factura_Diaria", palette="pastel", ax=axes[0])
axes[0].set_title("Factura Diaria según Evasión")
axes[0].set_xlabel("Evasión (0=No, 1=Sí)")
axes[0].set_ylabel("Factura Diaria")

# Boxplot Cantidad de Servicios
sns.boxplot(data=df_tl, x="Evasion", y="Cantidad_Servicios", palette="cool", ax=axes[1])
axes[1].set_title("Cantidad de Servicios vs Evasión")
axes[1].set_xlabel("Evasión (0=No, 1=Sí)")
axes[1].set_ylabel("Cantidad de Servicios")

plt.tight_layout()
plt.show()
```

**Introducción**

El presente informe presenta un análisis exploratorio de datos (EDA) para identificar los factores que influyen en la evasión de clientes (Churn) en la compañía Telecom X. Su objetivo es proporcionar una comprensión clara de los perfiles de clientes en riesgo y ofrecer recomendaciones estratégicas para mejorar la retención.

La pérdida de clientes es un problema crítico en el sector de las telecomunicaciones, ya que afecta de manera directa los ingresos y la rentabilidad de la empresa. Además, el costo de adquirir nuevos usuarios suele ser considerablemente mayor que el de mantener a los ya existentes, lo que refuerza la importancia de implementar estrategias preventivas.

A través de este análisis, se busca:

* Detectar patrones que distingan a los clientes que permanecen de aquellos que abandonan el servicio.

* Identificar variables clave que influyen en la decisión de cancelación.

* Generar información práctica que sirva de base para diseñar programas de retención más efectivos.

Este análisis constituye un primer paso para que el equipo de Data Science de Telecom X pueda avanzar en la creación de modelos predictivos y desarrollar estrategias que permitan reducir la evasión al anticiparse a la fuga de clientes y así actuar de forma proactiva para conservar su base de usuarios.


**Conclusiones y Hallazgos Principales**

La distribución de la variable de evasión `(Churn)` revela que, si bien la empresa retiene a la gran mayoría de sus clientes (71.2%), la tasa de evasión es lo suficientemente significativa (25.7%) como para requerir una intervención estratégica. Es importante destacar que un 3.1% de los datos en esta variable son valores nulos `(NaN)`, una inconsistencia que debe ser tratada antes de cualquier modelado predictivo para evitar sesgos.

**Factores Clave de la Evasión**

1. Antigüedad y Tipo de Contrato:

* El factor de riesgo más fuerte es el tipo de contrato. Los clientes con contratos mes a mes tienen una tasa de evasión drásticamente mayor en comparación con los contratos de uno o dos años (gráfico "Evasión por Tipo_Contrato")

* Existe una correlación negativa moderada entre la evasión y la antigüedad (`tenure`), lo que confirma que los clientes nuevos o con poca antigüedad son los más propensos a irse (gráfico de correlación y la distribución en el boxplot de tenure).

2. Servicios y Facturación:

* La evasión es significativamente más alta en clientes con Fibra Óptica (gráfico "Evasión por Servicio_Internet"). Esto sugiere un problema subyacente con la calidad del servicio, el soporte o el valor percibido.

* Los clientes que no tienen servicios adicionales de internet `(como seguridad, respaldo o soporte técnico)` son más propensos a evadir. Esto se complementa con el gráfico "Cantidad de Servicios vs Evasión", que muestra que los clientes que evaden tienen una menor cantidad de servicios. Los servicios adicionales actúan como un factor de retención.

* Los clientes que evaden tienden a tener facturas mensuales más altas. La combinación de facturas elevadas y pocos servicios contratados es un indicador de que el cliente percibe un bajo valor. El boxplot "Factura Diaria según Evasión" muestra que la mediana de la factura diaria es mayor para los clientes que evaden (1.0).

3. Método de Pago:

* El método de pago cheque electrónico está fuertemente asociado con una tasa de evasión notablemente más alta (gráfico "Evasión por Metodo_Pago"). Los métodos de pago automáticos (tarjeta de crédito o transferencia bancaria) actúan como una barrera de salida.

**Factores No Concluyentes**

* Las variables demográficas como el género, la pareja y los dependientes no muestran una relación significativa con la evasión (gráficos de barras correspondientes). La tasa de evasión es similar en estos grupos.

* Aunque los adultos mayores tienen una tasa de evasión proporcionalmente más alta, el impacto es menor en términos absolutos debido a que son un grupo más pequeño (gráfico "Evasión de clientes según Adulto Mayor").

**Recomendaciones Estratégicas para la Retención**

Basado en los hallazgos, se proponen las siguientes recomendaciones para reducir la evasión de clientes:

* Promoción de Contratos a Largo Plazo:

Implementar una estrategia agresiva para convertir a los clientes de contratos mes a mes en contratos de un año o dos años. Esto podría incluir descuentos especiales, beneficios exclusivos o tarifas congeladas para incentivar la permanencia.

* Evaluación y Mejora del Servicio de Fibra Óptica:

Realizar una investigación de la experiencia del cliente (encuestas, entrevistas) para identificar las causas exactas de la alta evasión en el servicio de fibra óptica. Abordar proactivamente los problemas de calidad, estabilidad, soporte o relación calidad-precio.

* Incentivar la Adopción de Servicios Adicionales:

Crear paquetes de servicios atractivos o promociones cruzadas para fomentar la suscripción a servicios adicionales de internet `(seguridad en línea, soporte técnico`, etc.). Estos servicios no solo aumentan los ingresos, sino que también mejoran la retención.

* Optimización de Métodos de Pago:

Fomentar la adopción de métodos de pago automáticos (tarjeta de crédito o transferencia bancaria). Ofrecer un pequeño descuento o un incentivo promocional para los clientes que se cambien de cheque electrónico a un método de pago automático.

* Programa de Retención para Clientes Nuevos:

Implementar un programa de bienvenida proactivo para los clientes con baja antigüedad (`tenure`). Esto podría incluir un seguimiento personalizado, un chequeo de satisfacción en los primeros 3 meses y un equipo de soporte dedicado para resolver rápidamente cualquier problema.

* Monitoreo y Personalización:

Establecer un sistema de monitoreo continuo para identificar a los clientes en riesgo (por ejemplo, aquellos con contratos mes a mes, facturas altas y pocos servicios). Utilizar esta información para segmentar y ofrecer ofertas personalizadas de forma proactiva antes de que decidan irse.

**Imputación de Valores Faltantes en la Evasión**

Para abordar los valores faltantes en la variable objetivo, Evasion, se implementó una estrategia de imputación avanzada para evitar el sesgo de rellenar los datos con un solo valor. En lugar de eliminar las filas o imputar un valor predeterminado, se utilizó el IterativeImputer, una técnica de imputación multivariada. Este método modela cada variable con valores faltantes como una función de otras variables en el conjunto de datos. En esencia, entrena un modelo predictivo (por defecto, un BayesianRidge) que toma las demás columnas como características para predecir los valores nulos en Evasion.

**Manejo de la Columna Factura_Total**

La columna Factura_Total presentaba una pequeña cantidad de valores nulos (NaN). Para manejar este problema, se optó por la imputación con la mediana en lugar de la media. La mediana es el valor central de una serie de datos ordenada y es significativamente más robusta ante la presencia de valores atípicos (outliers) que la media.
