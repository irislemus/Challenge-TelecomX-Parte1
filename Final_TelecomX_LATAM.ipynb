##Manejo de inconsistencias

```python
# Copia del dataframe para no modificar el original
df_tl = df_norm.copy()
```

```python
# Eliminar columnas no Ãºtiles
df_tl = df_tl.drop(columns=["customerID"])
```

```python
# Eliminar filas con Churn vacÃ­o
df_tl = df_tl.dropna(subset=['Churn'])
```

```python
# CodificaciÃ³n de columnas binarias
columnas_binarias = ["Churn", "Partner", "Dependents", "PhoneService", "PaperlessBilling", "gender"]
mapeo_binario = {"No": 0, "Yes": 1, "Male": 0, "Female": 1, "": np.nan}

for col in columnas_binarias:
    df_tl[col] = df_tl[col].map(mapeo_binario)
```

```python
# SeniorCitizen como booleano
df_tl["SeniorCitizen"] = df_tl["SeniorCitizen"].astype(bool)
```

```python
# Columnas multicategorÃ­a
columnas_multicategoricas = [
    "MultipleLines", "InternetService", "OnlineSecurity",
    "OnlineBackup", "DeviceProtection", "TechSupport",
    "StreamingTV", "StreamingMovies", "Contract", "PaymentMethod"
]

for col in columnas_multicategoricas:
    df_tl[col] = df_tl[col].astype("category")
```

```python
# ConversiÃ³n de columnas numÃ©ricas
df_tl["Total"] = pd.to_numeric(df_tl["Total"], errors="coerce")
df_tl['Total'] = df_tl['Total'].fillna(0.0)


df_tl["Monthly"] = pd.to_numeric(df_tl["Monthly"], errors="coerce")
df_tl["tenure"] = pd.to_numeric(df_tl["tenure"], errors="coerce")
```

RECUENTO DE EVASIÃ“N POR VARIABLES CATEGÃ“RICAS

```python
# para mantener lo hecho anteriormente en la transformaciÃ³n de los datos
df_plot = df_tl.copy()

# Mapear valores de gÃ©nero si son 0 y 1 para un mejor entendimient de la grÃ¡fica
if set(df_plot["Genero"].unique()) <= {0, 1}:
    df_plot["Genero"] = df_plot["Genero"].map({0: "Hombre", 1: "Mujer"})

# Variables categÃ³ricas a graficar
categoricas = [
    "Genero", "Pareja", "Dependientes",
    "Multiples_Lineas", "Servicio_Internet",
    "Tipo_Contrato", "Metodo_Pago"
]

# Calcular lÃ­mite mÃ¡ximo comÃºn
max_val = 0
for col in categoricas:
    conteos = df_plot.groupby([col, "Evasion"]).size()
    max_val = max(max_val, conteos.max())

# Crear subplots con mismo lÃ­mite
n_cols = 2
n_rows = math.ceil(len(categoricas) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(14, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(categoricas):
    sns.countplot(data=df_plot, x=col, hue="Evasion", palette="Set2", ax=axes[i])
    axes[i].set_title(f"EvasiÃ³n por {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Cantidad de Clientes")
    axes[i].legend(title="EvasiÃ³n", labels=["No", "SÃ­"])
    axes[i].tick_params(axis='x', rotation=45)
    axes[i].set_ylim(0, max_val * 1.1)  # mismo lÃ­mite para todos

# Ocultar ejes vacÃ­os si los hay
for j in range(len(categoricas), len(axes)):
    fig.delaxes(axes[j])

# Ajustar espaciado vertical y horizontal
plt.subplots_adjust(hspace=0.9, wspace=0.3)

plt.show()
```

**ImputaciÃ³n de Valores Faltantes en la EvasiÃ³n**

Para abordar los valores faltantes en la variable objetivo, Evasion, se implementÃ³ una estrategia de imputaciÃ³n avanzada para evitar el sesgo de rellenar los datos con un solo valor. En lugar de eliminar las filas o imputar un valor predeterminado, se utilizÃ³ el IterativeImputer, una tÃ©cnica de imputaciÃ³n multivariada. Este mÃ©todo modela cada variable con valores faltantes como una funciÃ³n de otras variables en el conjunto de datos. En esencia, entrena un modelo predictivo (por defecto, un BayesianRidge) que toma las demÃ¡s columnas como caracterÃ­sticas para predecir los valores nulos en Evasion.

**Manejo de la Columna Factura_Total**

La columna Factura_Total presentaba una pequeÃ±a cantidad de valores nulos (NaN). Para manejar este problema, se optÃ³ por la imputaciÃ³n con la mediana en lugar de la media. La mediana es el valor central de una serie de datos ordenada y es significativamente mÃ¡s robusta ante la presencia de valores atÃ­picos (outliers) que la media.
